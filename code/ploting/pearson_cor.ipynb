```python
import pandas as pd

basah_all = pd.read_csv('../raw_data/basah all.csv')
fermentasi_all = pd.read_csv('../raw_data/fermentasi all.csv')
kering_all = pd.read_csv('../raw_data/kering all.csv')
roasting_all = pd.read_csv('../raw_data/roasting all.csv')
```

```python
import seaborn as sns

import matplotlib.pyplot as plt

def plot_heatmap(df):
    corr = df.corr(method='pearson')
    plt.figure(figsize=(8, 6))
    sns.heatmap(corr, annot=True, cmap='coolwarm', fmt=".2f")
    plt.title('Pearson Correlation Heatmap')
    plt.show()
```

```python
plot_heatmap(basah_all)
plot_heatmap(fermentasi_all)
plot_heatmap(kering_all)
plot_heatmap(roasting_all)                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                                        
```

```python
def pearson_correlation_table(df):
    """
    Menghitung tabel korelasi Pearson dari dataframe pandas.

    Args:
        df (pd.DataFrame): Dataframe input.

    Returns:
        pd.DataFrame: Dataframe korelasi Pearson.
    """
    return df.corr(method='pearson')

# Contoh penggunaan:
pearson_correlation_table(basah_all)
```

```python
# Menghitung tabel korelasi Pearson untuk semua data dan menyimpannya ke file CSV
pearson_correlation_table(basah_all).to_csv('basah_all_pearson.csv')
pearson_correlation_table(fermentasi_all).to_csv('fermentasi_all_pearson.csv')
pearson_correlation_table(kering_all).to_csv('kering_all_pearson.csv')
pearson_correlation_table(roasting_all).to_csv('roasting_all_pearson.csv')
```
