```python
import pandas as pd

# Define file paths
gula_reduksi_file_path = '../curated_data_2/nir_basah_gula_reduksi_df.csv'
kadar_air_file_path = '../curated_data_2/nir_basah_kadar_air_df.csv'
lemak_file_path = '../curated_data_2/nir_basah_lemak_df.csv'
protein_file_path = '../curated_data_2/nir_basah_protein_df.csv'
fenol_file_path = '../curated_data_2/nir_basah_fenol_df.csv'


# Read the CSV files
gula_reduksi_df = pd.read_csv(gula_reduksi_file_path,index_col=0)
kadar_air_df = pd.read_csv(kadar_air_file_path,index_col=0)
lemak_df = pd.read_csv(lemak_file_path,index_col=0)
protein_df = pd.read_csv(protein_file_path,index_col=0)
fenol_df = pd.read_csv(fenol_file_path,index_col=0)
```

```python
# Drop the "KATEGORI" column from all relevant DataFrames if it exists
for df in [lemak_df, protein_df,gula_reduksi_df, kadar_air_df]:
    if "KATEGORI" in df.columns:
        df.drop(columns=["KATEGORI"], inplace=True)
    if "Kategori" in df.columns:
        df.drop(columns=['Kategori'], inplace=True)
```

```python
import sys
sys.path.append('../tools')
from preprossesing import mean_norm, max_norm, range_norm, snv, msc, sg1, sg2, smoothing_mean, fd1, fd2

# Drop 'PROTEIN' column for preprocessing
gula_reduksi_features = gula_reduksi_df.drop(columns=['Gula Reduksi'])
kadar_air_features = kadar_air_df.drop(columns=['Kadar Air'])
lemak_features = lemak_df.drop(columns=['LEMAK'])
protein_features = protein_df.drop(columns=['PROTEIN'])
fenol_features = fenol_df.drop(columns=['FENOL'])

# Apply preprocessing functions
gula_reduksi_mean_norm = mean_norm(gula_reduksi_features)
kadar_air_mean_norm = mean_norm(kadar_air_features)
lemak_mean_norm = mean_norm(lemak_features)
protein_mean_norm = mean_norm(protein_features)
fenol_mean_norm = mean_norm(fenol_features)

gula_reduksi_max_norm = max_norm(gula_reduksi_features)
kadar_air_max_norm = max_norm(kadar_air_features)
lemak_max_norm = max_norm(lemak_features)
protein_max_norm = max_norm(protein_features)
fenol_max_norm = max_norm(fenol_features)

gula_reduksi_range_norm = range_norm(gula_reduksi_features)
kadar_air_range_norm = range_norm(kadar_air_features)
lemak_range_norm = range_norm(lemak_features)
protein_range_norm = range_norm(protein_features)
fenol_range_norm = range_norm(fenol_features)

gula_reduksi_snv = snv(gula_reduksi_features)
kadar_air_snv = snv(kadar_air_features)
lemak_snv = snv(lemak_features)
protein_snv = snv(protein_features)
fenol_snv = snv(fenol_features)

gula_reduksi_msc = msc(gula_reduksi_features)
kadar_air_msc = msc(kadar_air_features)
lemak_msc = msc(lemak_features)
protein_msc = msc(protein_features)
fenol_msc = msc(fenol_features)

gula_reduksi_sg1 = sg1(gula_reduksi_features)
kadar_air_sg1 = sg1(kadar_air_features)
lemak_sg1 = sg1(lemak_features)
protein_sg1 = sg1(protein_features)
fenol_sg1 = sg1(fenol_features)

gula_reduksi_sg2 = sg2(gula_reduksi_features)
kadar_air_sg2 = sg2(kadar_air_features)
lemak_sg2 = sg2(lemak_features)
protein_sg2 = sg2(protein_features)
fenol_sg2 = sg2(fenol_features)

gula_reduksi_smoothing_mean = smoothing_mean(gula_reduksi_features,50)
kadar_air_smoothing_mean = smoothing_mean(kadar_air_features,50)
lemak_smoothing_mean = smoothing_mean(lemak_features,50)
protein_smoothing_mean = smoothing_mean(protein_features,50)
fenol_smoothing_mean = smoothing_mean(fenol_features,50)

gula_reduksi_fd1 = fd1(gula_reduksi_sg1)
kadar_air_fd1 = fd1(kadar_air_sg1)
lemak_fd1 = fd1(lemak_sg1)
protein_fd1 = fd1(protein_sg1)
fenol_fd1 = fd1(fenol_sg1)

gula_reduksi_fd2 = fd2(gula_reduksi_sg1)
kadar_air_fd2 = fd2(kadar_air_sg1)
lemak_fd2 = fd2(lemak_sg1)
protein_fd2 = fd2(protein_sg1)
fenol_fd2 = fd2(fenol_sg1)


# Add 'PROTEIN' column back to the processed data
gula_reduksi_mean_norm['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_mean_norm['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_mean_norm['LEMAK'] = lemak_df['LEMAK']
protein_mean_norm['PROTEIN'] = protein_df['PROTEIN']
fenol_mean_norm['FENOL'] = fenol_df['FENOL']

gula_reduksi_max_norm['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_max_norm['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_max_norm['LEMAK'] = lemak_df['LEMAK']
protein_max_norm['PROTEIN'] = protein_df['PROTEIN']
fenol_max_norm['FENOL'] = fenol_df['FENOL']

gula_reduksi_range_norm['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_range_norm['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_range_norm['LEMAK'] = lemak_df['LEMAK']
protein_range_norm['PROTEIN'] = protein_df['PROTEIN']
fenol_range_norm['FENOL'] = fenol_df['FENOL']

gula_reduksi_snv['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_snv['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_snv['LEMAK'] = lemak_df['LEMAK']
protein_snv['PROTEIN'] = protein_df['PROTEIN']
fenol_snv['FENOL'] = fenol_df['FENOL']

gula_reduksi_msc['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_msc['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_msc['LEMAK'] = lemak_df['LEMAK']
protein_msc['PROTEIN'] = protein_df['PROTEIN']
fenol_msc['FENOL'] = fenol_df['FENOL']

gula_reduksi_sg1['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_sg1['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_sg1['LEMAK'] = lemak_df['LEMAK']
protein_sg1['PROTEIN'] = protein_df['PROTEIN']
fenol_sg1['FENOL'] = fenol_df['FENOL']

gula_reduksi_sg2['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_sg2['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_sg2['LEMAK'] = lemak_df['LEMAK']
protein_sg2['PROTEIN'] = protein_df['PROTEIN']
fenol_sg2['FENOL'] = fenol_df['FENOL']

gula_reduksi_smoothing_mean['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_smoothing_mean['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_smoothing_mean['LEMAK'] = lemak_df['LEMAK']
protein_smoothing_mean['PROTEIN'] = protein_df['PROTEIN']
fenol_smoothing_mean['FENOL'] = fenol_df['FENOL']

gula_reduksi_fd1['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_fd1['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_fd1['LEMAK'] = lemak_df['LEMAK']
protein_fd1['PROTEIN'] = protein_df['PROTEIN']
fenol_fd1['FENOL'] = fenol_df['FENOL']

gula_reduksi_fd2['Gula Reduksi'] = gula_reduksi_df['Gula Reduksi']
kadar_air_fd2['Kadar Air'] = kadar_air_df['Kadar Air']
lemak_fd2['LEMAK'] = lemak_df['LEMAK']
protein_fd2['PROTEIN'] = protein_df['PROTEIN']
fenol_fd2['FENOL'] = fenol_df['FENOL']
```
